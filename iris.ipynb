#Data Visualization Using Seaborn

Iris DataSet:
It consists of meauremnts from 150 iris flowers from three different species, Setosa, Virginica, Versicolor.

For species we have measurements of sepal lenght, sepal width, petal lenght, petal width.The dataset is labelled as we know as species of each flower

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('/content/drive/MyDrive/Data Sets/Iris.csv')
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Species'].value_counts()
```

```python
sns.countplot(x='Species', data=df, palette='viridis')
plt.title('Counts of each species in Iris dataset')
```

```python
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm", data=df,hue="Species")
plt.figure(figsize=(8,6))
```

Insights:we can now distinguish the different species and examine their characteristics. This plot is helpful in understanding the featues with respect to sepal length and sepal width

Species Setosa has smaller sepal lengths but larger sepal widths.

Versicolor Species lies in the middle of the other two species in terms of sepal length and width

Species Virginica has larger sepal lengths but smaller sepal widths.

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df, )
```

Insights:Species Setosa has smaller petal lengths and widths.

Versicolor Species lies in the middle of the other two species in terms of petal length and width

Species Virginica has the largest of petal lengths and widths.

```python
sns.boxplot(x="Species", y="PetalLengthCm",data=df)
plt.figure(figsize=(10,8))
```

Insigths: This provides the information about the petal length of each species

```python
sns.violinplot(x="Species", y="SepalLengthCm",data=df)
plt.figure(figsize=(10,8))
```

```python
sns.kdeplot(data=df,x="PetalLengthCm", hue="Species")
```

Insights: This gives us the clear idea about the petal lenght in each species

```python
sns.pairplot(data=df,hue='Species')
plt.figure(figsize=(5,5))
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="Blues")
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.stripplot(x='Species', y='SepalLengthCm', data=df)
plt.subplot(2, 2, 2)
sns.stripplot(x='Species', y='SepalWidthCm', data=df, jitter=True, size=4)
```

```python
sns.swarmplot(x='Species', y = 'PetalLengthCm', data=df)
```

```python
sns.jointplot(x='SepalLengthCm', y = 'SepalWidthCm', data=df)
```
